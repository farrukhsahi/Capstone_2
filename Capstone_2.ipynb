```python
import pandas as pd
```

```python
# Total number of individuals
total_individuals = len(mh)

# Total number of unique countries
total_countries = mh['Country'].nunique()

print("Total number of individuals:", total_individuals)
print("Total number of countries:", total_countries)
```

```python
import matplotlib.pyplot as plt
```

```python
# Count the total number of males and females in the dataset
total_males = (mh['Gender'] == 'Male').sum()
total_females = (mh['Gender'] == 'Female').sum()

# Count the number of males and females who received treatment
males_treatment_yes = ((mh['Gender'] == 'Male') & (mh['treatment'] == 'Yes')).sum()
females_treatment_yes = ((mh['Gender'] == 'Female') & (mh['treatment'] == 'Yes')).sum()

# Create a pie chart for gender distribution
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.pie([total_males, total_females], labels=['Males', 'Females'], autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Males and Females in Dataset')

# Create a pie chart for treatment percentage
plt.subplot(1, 2, 2)
plt.pie([males_treatment_yes, females_treatment_yes], labels=['Male Treatment Yes', 'Female Treatment Yes'], autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Males and Females Who Received Treatment')
plt.tight_layout()
plt.show()
```

```python
# Calculating the total number of recorded cases grouped by gender
gender_total_cases = mh['Gender'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_total_cases, labels=gender_total_cases.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Recorded Cases by Gender')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Calculating the number of individuals seeking treatment grouped by gender
treatment_counts = mh.groupby('Gender')['treatment'].value_counts()

# Extracting the counts for 'Yes' treatment responses for each gender
yes_treatment_counts = treatment_counts[:, 'Yes']

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(yes_treatment_counts, labels=yes_treatment_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import geopandas as gpd

# Calculate the prevalence of mental health issues by country
country_counts = mh.groupby('Country')['treatment'].value_counts(normalize=True) * 100
country_prevalence = country_counts.loc[:, 'Yes'].reset_index(name='Prevalence')

# Load world map shapefile
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge world map with prevalence data
world = world.merge(country_prevalence, how='right', left_on='name', right_on='Country')

# Define continent centroids
continent_centroids = world.dissolve(by='continent')['geometry'].centroid

# Plot the world map
plt.figure(figsize=(15, 10))
world.plot(column='Prevalence', cmap='OrRd', linewidth=0.8, edgecolor='0.8', legend=True)
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add continent names as annotations
for continent, centroid in continent_centroids.items():
    plt.annotate(text=continent, xy=(centroid.x, centroid.y), xytext=(3, 3),
                 textcoords="offset points", fontsize=10, color='black')

plt.axis('off')
plt.show()
```

```python
# Group the data by occupation and count the total number of cases reported in each occupation
occupation_total_cases = mh['Occupation'].value_counts()

# Group the data by occupation and treatment, and count the number of individuals with treatment in each occupation
occupation_treatment_counts = mh[mh['treatment'] == 'Yes']['Occupation'].value_counts()

# Calculate the proportion of total cases reported by occupation
total_cases_proportion = (occupation_total_cases / len(mh)) * 100

# Calculate the proportion of cases reported by individuals who received treatment
treatment_cases_proportion = (occupation_treatment_counts / len(mh)) * 100

# Plot the bar chart
plt.figure(figsize=(12, 8))
plt.bar(total_cases_proportion.index, total_cases_proportion.values, color='skyblue', label='Total Cases (%)')
plt.bar(treatment_cases_proportion.index, treatment_cases_proportion.values, color='orange', label='Treatment (%)')
plt.xlabel('Occupation')
plt.ylabel('Percentage of Cases (%)')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Group the data by occupation and count the total number of cases reported in each occupation
occupation_total_cases = mh['Occupation'].value_counts()

# Calculate the proportion of total cases reported by occupation
total_cases_proportion = (occupation_total_cases / len(mh)) * 100

# Plot the bar chart
plt.figure(figsize=(12, 8))
plt.bar(total_cases_proportion.index, total_cases_proportion.values, color='lightgreen')
plt.xlabel('Occupation')
plt.ylabel('Proportion of Total Cases Reported (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculating the prevalence of mental health issues by occupation
occupation_counts = mh.groupby('Occupation')['treatment'].value_counts(normalize=True) * 100
occupation_prevalence = occupation_counts.loc[:, 'Yes'].reset_index(name='Prevalence')

# Plot the bar chart
plt.figure(figsize=(12, 8))
plt.bar(occupation_prevalence['Occupation'], occupation_prevalence['Prevalence'], color='skyblue')
plt.xlabel('Occupation')
plt.ylabel('Prevalence of Mental Health Issues (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculating the proportion of individuals seeking treatment by family history
family_history_counts = mh.groupby('family_history')['treatment'].value_counts(normalize=True) * 100
family_history_proportion = family_history_counts.unstack()

# Plot the stacked bar chart
plt.figure(figsize=(10, 6))
family_history_proportion.plot(kind='bar', stacked=True, color=['lightgreen', 'salmon'])
plt.xlabel('Family History of Mental Health Issues')
plt.ylabel('Proportion of Individuals (%)')
plt.xticks(rotation=0)
plt.legend(title='Treatment', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Define the order of the categories for the x-axis
category_order = ['1-14 days', '15-30 days', '31-60 days', 'More than 2 months', 'Go out Every day']

# Calculate the average stress level for each category of days spent indoors
stress_levels = mh.groupby('Days_Indoors')['Growing_Stress'].value_counts(normalize=True).unstack() * 100

# Remove the 'Maybe' column from the DataFrame
stress_levels.drop(columns=['Maybe'], inplace=True)

# Plot the line plot
plt.figure(figsize=(10, 6))
stress_levels.reindex(category_order).plot(marker='o')
plt.xlabel('Days Indoors')
plt.ylabel('Proportion of Individuals (%)')
plt.legend(title='Growing Stress', loc='lower left')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculating the frequency or proportion of individuals struggling with coping for each category of changes in habits
coping_struggles_counts = mh.groupby('Changes_Habits')['Coping_Struggles'].value_counts(normalize=True) * 100
coping_struggles_proportion = coping_struggles_counts.unstack()

# Plot the grouped bar chart
plt.figure(figsize=(10, 6))
coping_struggles_proportion.plot(kind='bar', stacked=False)
plt.xlabel('Changes in Habits')
plt.ylabel('Proportion of Individuals Struggling with Coping (%)')
plt.xticks(rotation=45)
plt.legend(title='Coping Struggles', loc='lower left')
plt.tight_layout()
plt.show()
```

```python
# Calculating the frequency or proportion of individuals with and without social weakness for each category of work interest
social_weakness_counts = mh.groupby('Work_Interest')['Social_Weakness'].value_counts(normalize=True) * 100
social_weakness_proportion = social_weakness_counts.unstack()

# Plot the grouped bar chart
plt.figure(figsize=(10, 6))
social_weakness_proportion.plot(kind='bar', stacked=False)
plt.xlabel('Work Interest')
plt.ylabel('Proportion of Individuals with Social Weakness (%)')
plt.xticks(rotation=45)
plt.legend(title='Social Weakness', loc='lower left')
plt.tight_layout()
plt.show()
```

```python
# Calculateing the frequency or proportion of individuals willing and unwilling to participate in a mental health interview for each category of mood swing intensity
interview_counts = mh.groupby('Mood_Swings')['mental_health_interview'].value_counts(normalize=True) * 100
interview_proportion = interview_counts.unstack()

# Plot the grouped bar chart
plt.figure(figsize=(10, 6))
interview_proportion.plot(kind='bar', stacked=False)
plt.xlabel('Mood Swings')
plt.ylabel('Proportion to Participate in Interview(%)')
plt.xticks(rotation=45)
plt.legend(title='Mental Health Interview', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Create a contingency table
contingency_table = pd.crosstab(mh['family_history'], mh['Mental_Health_History'])

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Individual Mental Health Issues')
plt.ylabel('Family History')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Convert timestamp column to datetime format
mh['Timestamp'] = pd.to_datetime(mh['Timestamp'])

# Extract year and month from the timestamp
mh['Year'] = mh['Timestamp'].dt.year
mh['Month'] = mh['Timestamp'].dt.month

# Group data by year and count the number of mental health issues per year
mental_health_yearly = mh.groupby('Year').size()

# Plot the temporal trends using a line plot
plt.figure(figsize=(10, 6))
mental_health_yearly.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Mental Health Issues')
plt.grid(True)
plt.xticks(mental_health_yearly.index)
plt.show()
```

```python
# Grouping the data by country and count the number of mental health issues
country_counts = mh.groupby('Country').size().reset_index(name='Mental_Health_Issues_Count')

# Sort the countries based on the count of mental health issues
country_counts_sorted = country_counts.sort_values(by='Mental_Health_Issues_Count', ascending=False)

# Select the top 5 countries
top_5_countries = country_counts_sorted.head(5)

print("Top 5 Countries with Mental Health Issues:")
print(top_5_countries)
```
